```python
from bs4 import BeautifulSoup
import requests
def scraper(url):
    return BeautifulSoup(requests.get(url).text,'html.parser')
# Gather data
wi = scraper('https://www.newrichmondwi.gov/?SEC=EDE32B8C-D13F-4AC1-8EB0-C3CED1C78D0F')
# Get relevant section of webpage
container = wi.find('div', class_='body')
# Get your urls
targets = [a['href'] for a in container.find_all('a', href=True)]
#Get only the pdf file names
[a.split('/')[-1] for a in targets]
```

```python
import pandas as pd
pd.DataFrame([a.split('/')[-1] for a in targets]).to_csv('whatchamacallit.csv')
```

```python
with open('whatchamacallit.csv','a') as fd:
    fd.write(pd.DataFrame([a.split('/')[-1] for a in targets]))
```
